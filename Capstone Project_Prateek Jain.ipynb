```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Uploading the primary data:
data = pd.read_excel('data.xlsx')
data.shape
```

```python
df=data.copy()
df.head()
```

```python
# Uploading the secondary data & merging into primary:
sec = pd.read_excel('Country-Code.xlsx')
df=pd.merge(df,sec,on='Country Code',how='left')
df.head()
```

## Importing, Understanding, and Inspecting Data:
    * Perform preliminary data inspection, findings the structure of the data, missing values, duplicates, etc.
    * Identify duplicates and remove them.

```python
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.info()
```

```python
# Identifying null values
df.isna().sum()
```

```python
df[df['Restaurant_Name'].isna()]
```

```python
# Dropping the null value of Restaurant Name, as there is only 1 such record. Also reseting the index
df.dropna(axis=0,subset=['Restaurant_Name'],inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df[df['Restaurant_Name'].isna()]
```

```python
df[df['Cuisines'].isna()]
```

```python
# We have 9 such records with no Cuinies, hence will replace them with 'Others'
df['Cuisines'].fillna('Others', inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

## Performing EDA

```python
# Geographical distribution of restaurants - By Countries
cntry_wise_res=df.groupby(['Country_Code','Country']).agg(Count = ('Restaurant_ID','count'))
cntry_wise_res.sort_values(by='Count', ascending = False)
```

```python
## With above table we can see that with most numbers of 8651 restaurants India is on top and on 2nd position is USA with 434
## restaurants and 3rd being UK with 80 restaurants.
```

```python
# Geographical distribution of restaurants - By Cities
City_wise_res=df.groupby(['Country','City']).agg(Count = ('Restaurant_ID','count'))
City_wise_res.sort_values(by='Count', ascending = False)
```

```python
## The above table shows that New Delhi, India has maximum restaurants with 5473, than 2nd Gurgaon, India with 1118 and 3rd 
## Noida, India with 1080. There are many cities accross countries where numnber of restaurants are only 1.
## Maximum number of restaurants = 5473, Minimum numbers of restaurants = 1
```

```python
Most_national_presence = df.groupby(['Restaurant_Name','Country']).Country.count()
Most_national_presence.sort_values(ascending = False)
```

```python
## The above table shows that 'Cafe Coffee Day' has the maximum presence with 83 outlets, 'Domino's Pizza' at 2nd with 79 
## outlets and 'Subway' at 3rd position with 63 outlets. All of these are present only in India.
```

```python
## Ratio between restaurants that allow table booking vs. those that do not allow table booking
Booking_table=df.groupby('Has_Table_booking').Restaurant_ID.count()
Booking_table
```

```python
print('Ratio between restaurants that allow table booking vs. those that do not allow table booking:', 
      round((Booking_table.Yes/Booking_table.No),2))
```

```python
# Graphical representation of above ration - Pie Chart
plt.title('Booking Availability')
plt.pie(Booking_table,labels=('Table Booking Not Allowed','Table Booking Allowed'),explode=(0.1,0),shadow=True,
        autopct='%1.2f%%')
plt.show()
```

```python
## Above pie chart clearly shows that only 12.13% restaurants allow table booking and majority of 87.87% don't allow.
```

```python
# Percentage of restaurants providing online delivery
Online_Delivery=df.groupby('Has_Online_delivery').Restaurant_ID.count()
Online_Delivery
```

```python
print('Percentage of restaurants providing online delivery:', round((Online_Delivery.Yes/Online_Delivery.sum()),4)*100)
```

```python
# Graphical representation via Pie Chart
plt.title('Online Delivery Availability')
plt.pie(Online_Delivery,labels=('Online Delivery Not Allowed','Online Delivery Allowed'),explode=(0.1,0),shadow=True,
        autopct='%1.2f%%')
plt.show()
```

```python
## The above calculation and the pie chart shows that 25.66% of restaurants provide online delivery. However, majority of 
## 74.34% do not have this facility
```

```python
# Difference in number of votes for the restaurants that deliver and the restaurants that do not deliver
Online_Delivery_Votes=df.groupby('Has_Online_delivery').Votes.sum()
Online_Delivery_Votes
```

```python
Votes_difference = Online_Delivery_Votes.No - Online_Delivery_Votes.Yes
print('Difference in number of votes for the restaurants that deliver and the restaurants that do not deliver:',
     Votes_difference)
```

```python
# Graphical representation via Pie Chart
plt.title('Online Delivery Votes Distribution')
plt.pie(Online_Delivery_Votes,labels=('No Online Delivery','Online Delivery'),explode=(0.1,0),shadow=True,
        autopct='%1.2f%%')
plt.show()
```

```python
## It means that 65.42% votes have been given to those restaurants who don't have online delivery and those with online delivery
## have received voting of 34.58%.
## It means that restaurants with no online delivery are getting more votes.
```

```python
Rating_corr=df[['Restaurant_ID','Restaurant_Name','Country','City','Aggregate_rating','Average_Cost_for_two','Votes',
                'Price_range','Has_Table_booking','Has_Online_delivery']]
```

```python
dummy = ['Has_Table_booking','Has_Online_delivery'] # 0 indicates 'NO' and 1 indicates 'YES'
Rating_corr = pd.get_dummies(Rating_corr,columns=dummy,drop_first=True)
Rating_corr = Rating_corr.merge(CuisineCount_by_res,left_on='Restaurant_Name',right_on='Restaurant_Name',how='left')
Rating_corr.head()
```

```python
# Correlation between "Rating" and "Average cost for two"
sns.regplot(x='Average_Cost_for_two',y='Aggregate_rating',data=Rating_corr)
Rating_corr[["Average_Cost_for_two", "Aggregate_rating"]].corr()
```

```python
## The above charts shows a weak positive correlation between Average cost for two and Rating of the restaurants
```

```python
# Correlation between "Rating" and "Has Table Booking" - 0 indicates 'NO' and 1 indicates 'YES'
sns.regplot(x='Has_Table_booking_Yes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Has_Table_booking_Yes','Aggregate_rating']].corr()
```

```python
## There is a good positive correlation between table booking option and Rating. This could be one of the basis for good rating 
```

```python
# Correlation between "Rating" and "Has Online delivery" - 0 indicates 'NO' and 1 indicates 'YES'
sns.regplot(x='Has_Online_delivery_Yes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Has_Online_delivery_Yes','Aggregate_rating']].corr()
```

```python
## There is a good positive correlation between online delivry option and Rating. This could be another big factor for giving a
## good rating to a restaurant
```

```python
# Correlation between "Rating" and "Price Range"
sns.regplot(x='Price_range',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Price_range','Aggregate_rating']].corr()
```

```python
## Price range has very positive correlation with rating, as with increasing price range rating also gets better. This clearly
## is a major factor for deciding the rating.
```

```python
# Correlation between "Rating" and "Votes"
sns.regplot(x='Votes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Votes','Aggregate_rating']].corr()
```

```python
## Similar to price Votes also have a very positive correlation with rating, as with increasing price range rating also gets 
## better. This is also a major factor for deciding the rating.
```

```python
# Correlation between "Rating" and "Cuisines Count"
sns.regplot(x='Cuisines',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Cuisines','Aggregate_rating']].corr()
```

```python
## Number of Cuisines provided by a restaurant is worst correlated to Rating. There is very minimal relation. Hence this cannot 
## be the deciding factor for Rating.
```
